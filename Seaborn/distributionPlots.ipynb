```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import seaborn as sns
```

```python
df = pd.read_csv('dm_office_sales.csv')
```

```python
df.head()
```

```python
sns.rugplot(x='salary',data=df,height=0.5)
```

```python
# Do not use displot
# sns.distplot()
# sns.histplot()

sns.displot(data=df,x='salary')
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=20,color='blue',edgecolor='black')
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=20,color='blue',edgecolor='black',linewidth=3,linestyle='dashed')
```

```python
sns.set(style='darkgrid')
sns.histplot(data=df,x='salary',bins=20,color='blue',edgecolor='black')
# Histplot and displot are almost same
```

```python
sns.set(style='darkgrid')
sns.displot(data=df,x='salary',bins=20,color='blue',edgecolor='black',kde='true',rug='true')
```

```python
sns.kdeplot(data=df,x='salary')
```

```python
np.random.seed(42)
sample_ages = np.random.randint(0,100,200)
```

```python
sample_ages
```

```python
sample_ages = pd.DataFrame(sample_ages,columns=['Age'])
sample_ages
```

```python
sns.rugplot(data=sample_ages,x='Age')
```

```python
sns.displot(data=sample_ages,x='Age',rug='true',bins=20)
```

```python
sns.displot(data=sample_ages,x='Age',rug='true',bins=20,kde=True)
```

```python
sns.kdeplot(data=sample_ages,x='Age')
# Now See in the kdeplot the precdiction goes from -20 to 120 ages as it gives a wider range than the actual
```

```python
sns.kdeplot(data=sample_ages,x='Age',clip=[0,100])
```

```python
sns.kdeplot(data=sample_ages,x='Age',clip=[0,100],bw_adjust=0.1,shade=True)
```
