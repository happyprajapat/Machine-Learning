```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
df = pd.read_csv("StudentsPerformance.csv")
```

```python
df.head()
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df, x='math score')
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df, y='math score', x='test preparation course')
```

```python
plt.figure(figsize=(10,4))
sns.boxplot(data=df, y='math score', hue='test preparation course', x="parental level of education")

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=df, x='math score', hue='test preparation course', y="parental level of education",palette='Set2')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=df, x='math score', y="parental level of education",palette='Set2')

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
plt.figure(figsize=(10,7))
sns.violinplot(data=df, x='math score', y="parental level of education",palette='Set2',hue='test preparation course')

plt.legend(bbox_to_anchor=(1.2,0.5))
```

```python
plt.figure(figsize=(10,7))
sns.violinplot(data=df, x='math score', y="parental level of education",palette='Set2',hue='test preparation course',split=True)

plt.legend(bbox_to_anchor=(1.2,0.5))
```

```python
plt.figure(figsize=(10,7))
sns.violinplot(data=df, x='math score', y="parental level of education",palette='Set2',split=True,inner='quartile')

plt.legend(bbox_to_anchor=(1.2,0.5))
```

```python
plt.figure(figsize=(10,7))
sns.violinplot(data=df, x='math score', y="parental level of education",palette='Set2',split=True,inner='stick')

plt.legend(bbox_to_anchor=(1.2,0.5))
```

```python
plt.figure(figsize=(10,7))
sns.violinplot(data=df, x='math score', y="parental level of education",palette='Set2',bw=0.1)

plt.legend(bbox_to_anchor=(1.2,0.5))
```

```python
sns.swarmplot(data=df, x='math score',y='gender',hue='parental level of education',size=3)

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.swarmplot(data=df, x='math score',y='gender',hue='parental level of education',size=3,dodge=True)

plt.legend(bbox_to_anchor=(1.05,0.5))
```

```python
sns.boxenplot(x='math score', y='test preparation course', data=df,hue='gender')
```
