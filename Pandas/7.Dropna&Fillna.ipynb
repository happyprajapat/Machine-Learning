dropna & fillna

```python
import pandas as pd
```

```python
var = pd.read_csv("FigureS4data.csv")
var
```

```python
var.dropna()
# Drops everyrow which cotains empty block/NaN
```

```python
var.dropna(axis=1)
```

```python
var.dropna(how="all")

# Removes the row with is fully empty
```

```python
var.dropna(subset=["A"])
```

```python
var.dropna(inplace=True)
var
```

```python
var.dropna(thresh=1)
# The row with only 1 NaN will be removed
```
