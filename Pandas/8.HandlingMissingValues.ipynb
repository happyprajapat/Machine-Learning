Replace

```python
import pandas as pd

var= pd.read_csv("FigureS4data.csv")
var
```

```python
var.replace(to_replace=1,value=2.2222222)
```

```python
var.replace([0.149388,2,3,4,5,6,7,8],"++")
var
```

```python
var.replace("[a-zA-z]","Happy",regex=True)
```

```python
var.replace({"D":"[A-Z]"},23232.3232,regex=True)
```

```python
var.replace(11,method="bfill")
```

```python
var.replace(1.000000,method="ffill",limit=3)
```
