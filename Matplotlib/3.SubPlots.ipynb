```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.style.use('fivethirtyeight')

nums = np.arange(20)
plt.figure(figsize=(15,4))
plt.suptitle("Graphs of degree 1, 2, 3",fontsize=24)


plt.subplot(1,3,1)
plt.plot(nums,nums)
plt.title("X")

plt.subplot(1,3,2)
plt.plot(nums,nums*nums,color='red')
plt.title("X Squared")


plt.subplot(1,3,3)
plt.plot(nums,nums**3,color='yellow')
plt.title("X Cubed")
plt.tight_layout()
```

```python
nums = np.arange(20)
plt.figure(figsize=(4,14))
plt.suptitle("Graphs of degree 1, 2, 3",fontsize=24)


plt.subplot(3,1,1)
plt.plot(nums,nums)
plt.title("X")

plt.subplot(3,1,2)
plt.plot(nums,nums*nums,color='red')
plt.title("X Squared")


plt.subplot(3,1,3)
plt.plot(nums,nums**3,color='yellow')
plt.title("X Cubed")
plt.tight_layout()
```
