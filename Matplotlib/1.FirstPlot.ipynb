Adjusting the figSize And Plot Dimensions

```python
plt.figure(figsize=(10,6)) #The size is in inches 
plt.plot(nums,nums)
```

```python
plt.figure(figsize=(2,6),dpi=200)
# dpi-> dots per inches
plt.plot(nums,nums)
```

```python
plt.figure(figsize=(6,6))
plt.plot(nums,nums**3)
```

Plot labels and titles

```python
agess=[12,20,24,28,30,35,50]
salaryy=[10000,25000,30000,40000,80000,120000,150000]

plt.figure(figsize=(3,3))

plt.plot(agess,salaryy)
plt.title("Salary Ages Graph")
plt.xlabel("Employee Age",labelpad=10)
plt.ylabel("Employee Salary",labelpad=10)

plt.figure()

plt.plot(nums,nums,color="blue",linewidth=4,linestyle="dashed")
plt.title("Linear Graph", loc="right")
plt.xlabel("Numbers",labelpad=10,fontsize=24)
plt.ylabel("Numbers",labelpad=10)
```

Changing xticks and yticks

```python
agess=[12,20,24,28,30,35,50]
salaryy=[10000,25000,30000,40000,80000,120000,150000]

plt.figure(figsize=(3,3))

plt.plot(agess,salaryy)
plt.title("Salary Ages Graph")
plt.xlabel("Employee Age",labelpad=10)
plt.ylabel("Employee Salary",labelpad=10)
plt.xticks([12,34,40,45,50])
plt.yticks([30000,50000,100000],labels=['30K','50K','100K'])
```

```python
agess=[20,24,28,30,35,50]
salaryy=[25000,30000,40000,80000,120000,150000]

plt.figure(figsize=(3,3))

plt.plot(agess,salaryy)
plt.title("Salary Ages Graph")
plt.xlabel("Employee Age",labelpad=10)
plt.ylabel("Employee Salary",labelpad=10)
plt.xticks([23,34,40,45,50])
plt.yticks([30000,50000,100000],labels=['30K','50K','100K'])

plt.xlim()
plt.xlim(30,40)
plt.ylim(50000,100000)
```

Adding Legends to plots

```python
plt.figure(figsize=(8,8))
plt.plot(nums,color="Green",label="x")
plt.plot(nums*nums,color="red",label="x-squared")
plt.plot(nums**3,color="yellow",label="x-cubed")
plt.legend()
```

```python
plt.figure(figsize=(8,8))
plt.plot(nums,color="Green",label="x")
plt.plot(nums*nums,color="red",label="x-squared")
plt.plot(nums**3,color="yellow",label="x-cubed")
plt.legend(loc="center left",frameon=True,shadow=True,facecolor="black")
```
