```python
import matplotlib.pyplot as plt
```

```python
plants=['Aloo','Gajar','Pyaaz','Gobi','Soyabean']
died=[10,25,5,30,21]
germinated=[74,88,56,69,59]
```

```python
plt.figure(figsize=(6,6))
# SYNTAX -> plt.bar(values,heights)
plt.bar(plants,germinated,width=0.5,color="pink")
plt.title("Germination Rates")
plt.ylabel("Plants Germinated")

```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,width=0.4,color='purple')
plt.title("Diying Rate")
plt.ylabel("Plants Died")

plt.figure()

plt.bar(plants,germinated,width=0.5,color="pink", align="edge")
plt.title("Germination Rates")
plt.ylabel("Plants Germinated")

```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,died,width=0.4,color='purple')
plt.bar(plants,germinated,width=0.5,color="pink", align="edge")


```

```python
plt.figure(figsize=(6,6))
plt.bar(plants,germinated,width=0.5,color="pink",bottom=died,label="Germinated Seeds")
plt.bar(plants,died,width=0.5,color='purple',label='Died Seeds')
plt.legend(shadow='True',frameon='True',facecolor='skyblue')
plt.title("All Seedlings")

```

```python
plt.figure(figsize=(6,6))
plt.barh(plants,died,color='purple')

```

Pie Charts

```python
names=['Turkey','Potatoes','Pimpkin seeds','Suffing']
prices=[32.1,23,45.2,65.8]
```

```python
plt.pie(prices)
```

```python
plt.pie(prices,labels=names)
```

```python
plt.pie(prices,labels=names, autopct='%1.1f%%')
plt.show()
```

```python
plt.pie(prices,labels=names, autopct='%1.1f%%',shadow='True')
plt.show()
```

```python
plt.pie(prices,labels=names, autopct='%1.0f%%',shadow='True',explode=(0,0.1,0,0))
plt.show()
```
